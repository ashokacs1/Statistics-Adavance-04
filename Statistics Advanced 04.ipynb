# Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.

A t-test and a z-test are both statistical tests used to make inferences about population parameters based on sample data. The key difference between these tests lies in the assumptions about the population and the sample size:

1. T-Test:
   - The t-test is used when the sample size is small (typically less than 30) and the population standard deviation is unknown.
   - It is suitable for testing hypotheses about a population mean.
   - The t-test uses the t-distribution, which has heavier tails than the standard normal (z) distribution, making it more appropriate for small sample sizes.
   - There are two common types of t-tests: the one-sample t-test and the two-sample t-test.
   
   Example Scenario: You want to test if a new drug has a statistically significant effect on reducing blood pressure in a small group of 20 patients. You don't know the population standard deviation, so you use a one-sample t-test to compare the sample mean to a known or hypothesized population mean.

2. Z-Test:
   - The z-test is used when the sample size is large (typically greater than 30) or when the population standard deviation is known.
   - It is also used to test hypotheses about a population mean.
   - The z-test uses the standard normal distribution, which is a normal distribution with a mean of 0 and a standard deviation of 1.
   
   Example Scenario: You are studying the heights of adult males in a large city and have access to the population's standard deviation. You want to determine if the average height of males in your sample of 100 individuals is significantly different from the known population mean. In this case, you would use a z-test.

In summary, the choice between a t-test and a z-test depends on the sample size and whether you know the population standard deviation. If the sample size is small or the population standard deviation is unknown, you should use a t-test. If the sample size is large and the population standard deviation is known, a z-test is more appropriate.

# Q2: Differentiate between one-tailed and two-tailed tests.

One-Tailed Test and Two-Tailed Test are two different types of hypothesis tests in statistics that are used to determine the significance of an effect or difference. The key difference between them lies in how they define the critical region and what kind of alternative hypotheses they test:

1. One-Tailed Test:
   - A one-tailed test, also known as a directional test, is used when you have a specific direction in mind for your hypothesis. You are interested in whether a parameter is significantly greater than or less than a certain value.
   - In a one-tailed test, the critical region is located in only one tail of the distribution (either the right tail or the left tail, depending on the direction specified in the alternative hypothesis).
   - The null hypothesis (H0) represents the status quo, while the alternative hypothesis (Ha) specifies a directional change.

   Example: You want to test if a new weight loss program is more effective in helping individuals lose weight compared to the current program. Your null hypothesis (H0) could be that there is no difference or the new program is less effective, and the alternative hypothesis (Ha) would state that the new program is more effective.

2. Two-Tailed Test:
   - A two-tailed test, also known as a non-directional test, is used when you are interested in whether a parameter is significantly different from a certain value, but you don't have a specific direction in mind. You are looking for any significant difference, whether it's greater or less than the specified value.
   - In a two-tailed test, the critical region is split between both tails of the distribution.
   - The null hypothesis (H0) typically states that there is no significant difference, and the alternative hypothesis (Ha) states that there is a significant difference, without specifying a particular direction.

   Example: You want to test if the mean IQ of a group of students is different from the national average IQ of 100. In this case, the null hypothesis (H0) would be that there is no difference, and the alternative hypothesis (Ha) would be that there is a difference, without specifying whether it's higher or lower.

In summary, the choice between a one-tailed test and a two-tailed test depends on your research question and whether you have a specific directional hypothesis in mind. A one-tailed test is appropriate when you are interested in a specific direction of change, while a two-tailed test is used when you are only interested in detecting any significant difference, regardless of the direction.

# Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.

In hypothesis testing, there are two types of errors that can occur: Type I error and Type II error. These errors are associated with the incorrect rejection or acceptance of a null hypothesis when conducting a statistical test.

1. Type I Error (False Positive):
   - Type I error occurs when you incorrectly reject a null hypothesis that is actually true. In other words, you conclude that there is a significant effect or difference when, in reality, there is none.
   - The probability of making a Type I error is denoted as the significance level (alpha, α), and it is typically set by the researcher (e.g., 0.05 for a 5% significance level).
   - Type I error is also known as a "false positive" or a "false alarm."

   Example Scenario: A pharmaceutical company is testing a new drug for a rare disease. The null hypothesis (H0) is that the drug has no effect on the disease. If, in reality, the drug has no effect (H0 is true), but the statistical test falsely indicates that the drug is effective (rejects H0), it would be a Type I error.

2. Type II Error (False Negative):
   - Type II error occurs when you incorrectly fail to reject a null hypothesis that is actually false. In other words, you conclude that there is no significant effect or difference when, in reality, there is one.
   - The probability of making a Type II error is denoted as beta (β).
   - Type II error is also known as a "missed opportunity" or a "false negative."

   Example Scenario: In a quality control process, the null hypothesis (H0) is that a manufacturing process is producing products with a certain level of defects. If, in reality, the process is producing more defective products (H0 is false), but the statistical test fails to detect this and does not reject H0, it would be a Type II error.

In summary, the concepts of Type I and Type II errors are crucial in hypothesis testing because they highlight the trade-off between the risks of making incorrect decisions. Researchers aim to balance the two types of errors based on their specific goals and the consequences of each error type in a given scenario. Adjusting the significance level (α) and sample size can impact the likelihood of Type I and Type II errors, with a trade-off between them: reducing the probability of one type of error generally increases the probability of the other.

# Q4: Explain Bayes's theorem with an example.

Bayes's theorem is a fundamental concept in probability theory and statistics that describes how to update the probability of an event based on new information. It relates conditional probabilities, which are the probabilities of an event occurring given that another event has already occurred. The formula for Bayes's theorem is as follows:

\[P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}\]

Where:
- \(P(A|B)\) is the conditional probability of event A occurring given that event B has occurred.
- \(P(B|A)\) is the conditional probability of event B occurring given that event A has occurred.
- \(P(A)\) is the prior probability of event A occurring (before considering the new information).
- \(P(B)\) is the prior probability of event B occurring (before considering the new information).

Bayes's theorem is particularly useful in situations where you want to update your belief or knowledge about an event when new evidence becomes available. It's commonly used in Bayesian statistics, machine learning, and various real-world applications, including medical diagnosis, spam email filtering, and more.

Let's illustrate Bayes's theorem with an example:

Example: Medical Diagnosis

Suppose you are a doctor and you need to diagnose a rare disease, "Disease X," in a patient. The disease affects only 1% of the population. You have two diagnostic tests available:

- Test A has a sensitivity of 90%, which means it correctly identifies the disease in 90% of the cases where it's present but gives false positives in 10% of the cases where the disease is not present.
- Test B has a sensitivity of 80% and specificity of 95%, meaning it correctly identifies the disease in 80% of the cases where it's present and correctly identifies its absence in 95% of the cases where it's not present.

You have a new patient who tests positive for both Test A and Test B. You want to determine the probability that the patient actually has Disease X.

Using Bayes's theorem, you can calculate this as follows:

- \(P(A)\): Prior probability of having Disease X = 1% = 0.01
- \(P(B|A)\): Probability of testing positive for Test B given having Disease X = Sensitivity of Test B = 80% = 0.80
- \(P(B|\neg A)\): Probability of testing positive for Test B given not having Disease X = False positive rate of Test B = 5% = 0.05

Now, you can apply Bayes's theorem:

\[P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}\]

\[P(B) = P(B|A) \cdot P(A) + P(B|\neg A) \cdot P(\neg A)\]

\[P(\neg A) = 1 - P(A) = 1 - 0.01 = 0.99\]

Now, calculate \(P(B)\):

\[P(B) = (0.80 \cdot 0.01) + (0.05 \cdot 0.99) = 0.008 + 0.0495 = 0.0575\]

Finally, calculate \(P(A|B)\):

\[P(A|B) = \frac{0.80 \cdot 0.01}{0.0575} ≈ 0.1391\]

So, based on the positive results from both tests, the patient has approximately a 13.91% chance of actually having Disease X. Bayes's theorem helps in updating the probability of the patient having the disease given the new evidence from both tests.

# Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

A confidence interval is a range of values that provides a plausible range for an unknown population parameter, such as a population mean or proportion, based on a sample from that population. It is a way to quantify the uncertainty or variability in your estimate of the parameter. Confidence intervals are commonly used in statistics to convey the level of confidence in the accuracy of your estimate.

A confidence interval is typically expressed in the form of "point estimate ± margin of error." The "point estimate" is the sample statistic (e.g., the sample mean or proportion), and the "margin of error" indicates the range within which the true population parameter is likely to fall with a specified level of confidence.

To calculate a confidence interval, you need the following information:

1. Sample data: You should have a sample from the population of interest.

2. Confidence level (usually denoted as \(1 - \alpha\)): This represents the desired level of confidence in your estimate. Common confidence levels include 90%, 95%, and 99%.

3. Standard error: The standard error is a measure of the variability in the sample statistic and depends on the population parameter you are estimating. The standard error is calculated based on the sample data.

The formula for calculating a confidence interval for a population mean (\(\mu\)) is as follows:

\[\text{Confidence Interval} = \text{Sample Mean} ± \text{Margin of Error}\]

The margin of error is determined by the formula:

\[\text{Margin of Error} = \text{Critical Value} × \frac{\text{Standard Deviation}}{\sqrt{n}}\]

Where:
- "Critical Value" is the value from the standard normal distribution (z-table) corresponding to the desired confidence level (e.g., for a 95% confidence level, the critical value is approximately 1.96).
- "Standard Deviation" is the population standard deviation (if known) or the sample standard deviation.
- "n" is the sample size.

Let's illustrate this with an example:

Example: Confidence Interval for Mean

Suppose you want to estimate the average height of a population. You take a random sample of 100 people from the population and measure their heights. The sample mean height is 165 cm, and the sample standard deviation is 8 cm. You want to calculate a 95% confidence interval for the population mean height.

1. Find the critical value for a 95% confidence level. Using a standard normal distribution table or calculator, you find the critical value is approximately 1.96.

2. Calculate the standard error:

\[\text{Standard Error} = \frac{8}{\sqrt{100}} = 0.8\]

3. Calculate the margin of error:

\[\text{Margin of Error} = 1.96 × 0.8 = 1.568\]

4. Calculate the confidence interval:

\[\text{Confidence Interval} = 165 ± 1.568\]

So, the 95% confidence interval for the population mean height is \(165 ± 1.568\), which means you are 95% confident that the true average height of the population falls within the range of approximately 163.43 cm to 166.57 cm.

This interval provides a level of confidence that the true population mean height lies within this range, based on the sample data and the chosen confidence level.

- Alternative Hypothesis (\(H_a\)): Caffeine has a significant effect on reaction time, or the population mean reaction time with caffeine (\(\mu\)) is not equal to the baseline reaction time (\(0.25\) seconds). Mathematically, \(H_a: \mu \neq 0.25\).

**Step 2: Set the Significance Level (\(\alpha\)):**
We are given a 90% confidence level, which corresponds to a significance level of \(\alpha = 0.10\).

**Step 3: Collect and Analyze the Data:**
- Sample size (\(n\)) = 30
- Sample mean (\(\bar{x}\)) = 0.25 seconds
- Sample standard deviation (\(s\)) = 0.05 seconds

**Step 4: Calculate the Test Statistic:**
We will use a one-sample t-test for this analysis. The test statistic is calculated as follows:

\[
t = \frac{\bar{x} - \mu}{\frac{s}{\sqrt{n}}}
\]

Where:
- \(\bar{x}\) is the sample mean.
- \(\mu\) is the hypothesized population mean under the null hypothesis.
- \(s\) is the sample standard deviation.
- \(n\) is the sample size.

Substituting the values:

\[
t = \frac{0.25 - 0.25}{\frac{0.05}{\sqrt{30}}}
\]

Calculate the value of \(t\).

**Step 5: Determine the Critical Value:**
Since we want to conduct a two-tailed test and find the critical t-values for a 90% confidence level, we need to find the critical t-values for \(\alpha/2\) and degrees of freedom (\(df = n - 1\)). In this case, \(df = 30 - 1 = 29\). You can find the critical t-values from a t-table or calculator.

**Step 6: Make a Decision:**
Compare the calculated \(t\)-value from Step 4 to the critical \(t\)-values from Step 5. If the calculated \(t\)-value falls outside the range defined by the critical \(t\)-values, you will reject the null hypothesis. If it falls within the range, you will fail to reject the null hypothesis.
